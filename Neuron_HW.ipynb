### Training

```python
X_train = [
    [
        [5,-1, 0],
        [2, 0,-5],
        [1, 1,-5]
    ],
    [
        [ 0,-2, 1],
        [ 1,-1, 4],
        [-1, 5,-5]
    ],
    [
        [0,-4, 0],
        [2, 2, 0],
        [1, 1,-4]
    ],
    [
        [1,-1, 5],
        [1, 4, 5],
        [0, 2,-4]
    ],
    [
        [5,-1, 1],
        [5, 2, 5],
        [0, 2,-4]
    ],
    
]

Y_train = [
    1,
    0,
    0,
    0,
    0,
]
X_test = [
    [
        [0,-5, 0],
        [-1, 5, 1],
        [1, 1,-4]
    ],
    [
        [-1,-1, 5],
        [-4, 4, 1],
        [ 0, -1,-4]
    ],
    [
        [5,-1, 1],
        [1, 2, 5],
        [0,-5,-4]
    ],
]
Y_test = [
    0,
    0,
    0,
]
```

```python
class NeuronDetect:
    def __init__(self):
        self.weights = [
            [
                0 for j in range(3)
            ]  for i in range(3)
        ]
        self.bias    = 0
    def sigmoidActivation(self, x):
        e = 2.718281828459045
        return 1 / (1  +  e  **  -x)
    def forward(self, X):
        y = self.bias
        for i in range(3): 
            for j in range(3):
                y += X[i][j] * self.weights[i][j]
        
        y = self.sigmoidActivation(y)
        return y
    def __str__(self):
        out = ""
        for i in range(9):
            if i % 3 == 0:
                out += "\n"
            out += f"{self.weights[i]:10.5f}"
        out += "\n"
        out += f"{self.bias:10.5f}"
        return out

def loss(Y, Yp):
    return Y - Yp


def reshapeMatrix(X):
    Xt0 = []
    
    for j in range(3):
        Xt0.append( X[j][0] )
    Xt1 = []

    for j in range(3):
        Xt1.append( X[j][1] )
    Xt2 = []

    for j in range(3):
        Xt2.append( X[j][2] )
    Xt_Final = [
        Xt0,
        Xt1,
        Xt2,
    ]
    return Xt_Final
```

```python
neuron = NeuronDetect()
# print(neuron.weights)

lr = 0.01
for epoch in range(1,6000):
    error = []
    # print("------------------------------------------")
    # print(f"Epoch: {epoch}")
    # print()
    for i in range(5):
        # print(f"{i+1}: ", end = "")
        xt = X_train[i]
        yt = Y_train[i] 
        xt = reshapeMatrix(xt)
        yp = neuron.forward(xt)
        e = loss(yt, yp)
        dBias = e * lr
        neuron.bias += dBias
    
        for i in range(3):
            for j in range(3):
                dweight = lr * e * xt[i][j]
                neuron.weights[i][j] += dweight
                error.append(e)
        me = abs(sum(error) / len(error))
        
        # print(f"Yt: {yt:5} Yp: {yp:4.3} e: {me:5.4}")
        # print()


    
    
```
